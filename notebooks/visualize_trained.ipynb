```python
import numpy as np
import os
from matplotlib import pyplot as plt

def visualize(result_path, img_data, real_data, predict_data, predicted_data, idx):
    
    f = plt.figure(figsize = (8,8))
    
    f.add_subplot(2,2,1)
    cs = plt.imshow(img_data[idx,:,:,0])
    cbar = f.colorbar(cs)
    cbar.ax.minorticks_off()
    plt.title('image_c0')
    
    f.add_subplot(2,2,2)
    cs = plt.imshow(img_data[idx,:,:,1])
    cbar = f.colorbar(cs)
    cbar.ax.minorticks_off()
    plt.title('image_c1')
    
    f.add_subplot(2,2,3)
    cs = plt.imshow(real_data[idx,:,:])
    cbar = f.colorbar(cs)
    cbar.ax.minorticks_off()
    plt.title('groundtruth')
    
    f.add_subplot(2,2,4)
    cs = plt.imshow(predicted_data[idx,:,:])
    cbar = f.colorbar(cs)
    cbar.ax.minorticks_off()
    plt.title('prediction')	
    plt.show()
#     plt.savefig(result_path+'/mask_comparison_idx%d.png'%(idx))
    plt.close()
```

```python
results_path = '/home/yifanc3/Github/cell_division/results/tryout'
test_x = np.load(results_path +'/inputs.npy')
test_y = np.load(results_path +'/gt_labels.npy')
result_path = '/home/yifanc3/Github/cell_division/results/tryout/weights.38-results-test'
pred = np.load(result_path +'/pred_labels.npy')
predicted_data = (pred>0.5).astype('uint8')
```

```python
img = test_x
real = test_y

for i in range(100):
    visualize(result_path,img,real,pred,predicted_data,i)
```
